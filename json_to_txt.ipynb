```python
import json
```

```python
##
# parameters  nomFichierJson: le nom du fichier json qui contient les informations du STNU
#             nomFichierTxt : le nom du fichier txt dans lequel on va rentrer les informations du STNU
#             Si nomFichier n'est pas préciser alors il prend le même nom que nomFichierJson
# les STNU dans le dossier stnu ne sont pas aboutis on ne peut donc pas les utiliser ici
def create_txt_from_json(nomFichierJson, nomFichierTxt=""):

    if nomFichierTxt == "":
        nomFichierTxt = nomFichierJson
    
    # On récupère les informations du STNU contenu dans le ficheir json
    with open(f"{nomFichierJson}.json", "r", encoding = "utf-8") as g:
        stnu = json.load(g)

    with open(f"{nomFichierTxt}.txt", "w", encoding = "utf-8") as f:

        nodes = stnu["nodes"]
        edges = stnu["constraints"]
        f.write(f"{len(nodes)} {len(edges)}\n")

        uncontrollableNodes = []

        for node in nodes:
            nom = node["node_id"]
            type = node["type"]
            if type == "u":
                uncontrollableNodes.append((nom,type))
            else:
                f.write(f"{nom} {type}\n")

        for node in uncontrollableNodes:
            f.write(f"{node[0]} {node[1]}\n")

        controllableEdges = []

        for edge in edges:
            type = edge["type"]
            firstNode = edge["first_node"]
            secondNode = edge["second_node"]
            tMin = edge["min_duration"]
            tMax = edge["max_duration"]
            if type == "f":
                controllableEdges.append((type, firstNode, secondNode, tMin, tMax))
            else:
                f.write(f"1 {type} {firstNode} {secondNode} {tMin} {tMax}\n")
        
        for edge in controllableEdges:
            f.write(f"1 {edge[0]} {edge[1]} {edge[2]} {edge[3]} {edge[4]}\n")
```

```python
create_txt_from_json("Exemple_STNU", "Exemple_STNU")
```
