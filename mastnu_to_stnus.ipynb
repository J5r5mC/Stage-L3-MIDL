```python
import json
```

```python
def ecrire_stnu(donnees_agents):
    agents = list(donnees_agents.keys())
    for i in range(len(agents)):
        with open(f"stnu/{agents[i]}_stnu.json", "w", encoding = "utf-8") as f:
            json.dump(donnees_agents[f"{agents[i]}"], f, indent=4)

def transform_mastnu_in_stnus(nomFichierSTN):

    with open(f"{nomFichierSTN}.json", "r", encoding = "utf-8") as g:
        mastn = json.load(g)
    
    nodes = mastn["nodes"]
    agents_rencontres = []
    donnees_agents = {}

    for i in range(len(nodes)) :
        node = nodes[i]
        agents = node["agents"]

        for j in range(len(agents)) :

            agent = agents[j]

            if (not(agent in agents_rencontres)):
                agents_rencontres.append(agents[j])
                donnees_agents[f"{agent}"] = {"nodes" : [], "constraints" : []}

            if (len(agents) > 1):
                node_type = "u" 

            else: 
                node_type = "c"
            
            donnees_agent = donnees_agents[f"{agent}"]
            node_id = len(donnees_agent["nodes"])

            if (node["time_spec"] == "E_AT_START") : 
                description = "Start : " + node["action"]

            else :
                description = "End : " + node["action"]
                # for node_start in nodes:
                #     if node_start["action"] == node["action"] and node_start["time_spec"] == "E_AT_START":
                #         start_source = node_start
                # for node_start in donnees_agent["nodes"]:
                #     if node_start["descrption"] == "Start : " + start_source["action"]:
                #         start_destination = node_start 
                # first_node = start_destination["node_id"]
                # second_node = node_id
                # egde_type = node_type


            donnees_agent["nodes"].append({"node_id" : node_id,
                                          "type" : node_type,
                                          "description" : description})
            
    ecrire_stnu(donnees_agents)
            
            
           
```
